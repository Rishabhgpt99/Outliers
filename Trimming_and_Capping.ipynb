```python
import pandas as pd
```

```python
df=pd.read_csv('/content/weight-height (1).csv')
```

```python
df.head()
```

# Finding outliers by IQR

```python
#
Q1= df['Height'].quantile(0.25)
Q1
```

```python
Q3=df['Height'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit= Q1 - (1.5*IQR)
lower_limit
```

```python
upper_limit= Q3 + (1.5*IQR)
upper_limit
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['Height'], vert=False)
```

# Removing Outliers by Trimming process


```python
df1= df[df['Height']<upper_limit]
plt.boxplot(df1['Height'], vert=False)
```

```python
df1.shape
```

```python
df2=df1[df1['Height']>lower_limit]
plt.boxplot(df2['Height'], vert=False)
```

```python
df2.shape
```

By Capping Process..

```python
df['Height']=df['Height'].where(df['Height']<=upper_limit,upper_limit)
df['Height']=df['Height'].where(df['Height']>=lower_limit,lower_limit)
```
